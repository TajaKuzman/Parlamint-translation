```python
import pandas as pd
import os

# Define the language code, used in the file names
lang_code = "NO"

# Main path
main_path = "/home/tajak/Parlamint-translation"

# Define the translation model to be used
opus_lang_code = "sv"

# Check whether the path to the folder with conllu files is ok
path = "{}/Source-data/ParlaMint-{}.conllu/ParlaMint-{}.conllu".format(main_path, lang_code, lang_code)

# Define other paths
extracted_dataframe_path = "{}/results/{}/ParlaMint-{}-extracted-source-data.csv".format(main_path, lang_code, lang_code)

translated_dataframe_path = "{}/results/{}/ParlaMint-{}-translated.csv".format(main_path, lang_code, lang_code)
translated_tokenized_dataframe_path = "{}/results/{}/ParlaMint-{}-translated-tokenized.csv".format(main_path,lang_code, lang_code)
final_dataframe = "{}/results/{}/ParlaMint-{}-final-dataframe.csv".format(main_path,lang_code, lang_code)
```

# Extracted dataset

```python
df = pd.read_csv(extracted_dataframe_path, sep="\t", index_col = 0, na_filter = False)

df.describe(include="all")
```

```python
df["text"].fillna(value = " ", inplace = True)
df[df["text"] == " "].to_dict()
```

```python
df.length.sum()
```

```python
# Extract a list with paths to conllu files and a list with their names
parl_list = []
file_name_list = []

for dir1 in os.listdir(path):
    full_path = os.path.join(path, dir1)
    if os.path.isdir(full_path):
        current = os.listdir(full_path)
        # Keep only files with parliamentary sessions:
        for file in current:
            if "ParlaMint-{}_".format(lang_code) in file:
                if ".conllu" in file:
                    final_path = "{}/{}".format(full_path, file)
                    parl_list.append(final_path)
                    file_name_list.append(file)

print(parl_list[:2])
print(file_name_list[:2])

# See how many files we have:
len(parl_list)
```

```python
# Inspect which file is not in the dataframe, print it and print its length (check if it is empty)
for i in file_name_list:
    if i not in list(df.file.unique()):
        print(i)
        file = open(parl_list[file_name_list.index(i)], "r").read()
        print(file)
        print(len(file.split()))
```

```python
df = pd.read_csv(final_dataframe, sep="\t", index_col = 0)

df.head(1)
```

```python
# Analyze only instances with substitutions
# Show differences between translations that we will use and translations after post-processing
with pd.option_context('display.max_colwidth', None):
    display(df[df["substituted_pairs"] !="0"][['text','translation','new_translations', 'substituted_pairs']].tail(20))
```

```python
# In Bulgarian dataframe, we won't use the post-processed translations, because Bulgarian uses cyrillic script and the substitution with the original lemma looks like this: "Importer : иван станков ." So I will substitute the new_translations with the original ones and use the original translations in the next steps. The same process was done for languages where the substitutions would introduce more errors than solutions: Portuguese, Italian, Greek, Austrian, Hungarian.
df["post-processed_translations"] = df["new_translations"]

df["new_translations"] = df["translation-tokenized"]

df[df["substituted_pairs"] != "0"].head()
```

```python
# Save the new df
df.to_csv("{}".format(final_dataframe), sep="\t")
```
