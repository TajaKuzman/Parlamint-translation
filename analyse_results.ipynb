# Translated df

```python
df_translated = pd.read_csv(translated_tokenized_dataframe_path, sep="\t", index_col=0, na_filter=False)
df_translated.describe(include="all")
```

```python
# Add translation length
df_translated["tran_length"] = df_translated["translation"].str.split().str.len()
df_translated.head(2)
```

```python
# Add info where tran length is more than 3 times longer than original length
df_translated["shorten"] = np.where((df_translated["tran_length"] > 6) & (df_translated["tran_length"] > 4*df_translated["length"]), "yes", "no")

df_translated.head()

```

```python
df_translated[df_translated["shorten"] == "yes"][["text", "length", "translation", "tran_length"]]
```

```python
shorten_translation = []

for i in list(zip(df_translated["translation"], df_translated["shorten"], df_translated["length"])):
    if i[1] == "no":
        shorten_translation.append(i[0])
    elif i[1] == "yes":
        new_length = 3*i[2]
        new_translation_list = i[0].split()[:new_length]
        new_translation = " ".join(new_translation_list)
        shorten_translation.append(new_translation)
        print("Old translation, new translation")
        print(i[0])
        print(new_translation)

df_translated["shorten_translation"] = shorten_translation
df_translated.head
```
